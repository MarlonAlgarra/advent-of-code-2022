```python
with open("day04.txt", "r") as f:
    txt_list = [line.strip() for line in f]
txt_list
```

```python
counter = 0
for item in txt_list:
    tempRange = [i.split('-') for i in item.split(',')]
    stringList=[]
    for j in tempRange:
        tempString = ''
        for k in range(int(j[0]),int(j[1])+ 1):
            tempString += str(k)
        stringList.append(tempString)
    print(stringList)
    #if stringList[0] in stringList[1] or stringList[1] in stringList[0]:
    #    counter+=1
print(counter)
```
